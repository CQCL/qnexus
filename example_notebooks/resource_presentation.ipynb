```python
import qnexus as qnx
```

### A 'filter' method returns a NexusDatabaseIterator

```python
my_jobs_data = qnx.jobs.filter(name="compile", status=["ERROR"])
```

```python
# Can be summarized to get a brief overview of the matches

my_jobs_data.summarize()
```

### Calling all() on a PaginationIterator loads all the data and returns a RefList, which supports the summarize protocol

```python
my_jobs_data.all().summarize()
```
