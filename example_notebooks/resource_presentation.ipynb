```python
import qnexus as qnx
```

### A 'get' method returns a NexusDatabaseIterator

```python
my_jobs_data = qnx.job.get(name_like="compile", status=["ERROR"])
```

```python
# Can be summarized to get a brief overview of the matches

my_jobs_data.summarize()
```

### Calling list()loads all the data and returns a RefList

```python
my_jobs_data.list()
```

### Alternatively can be displayed in a dataframe

```python
my_jobs_data.df()
```
